```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
output_fname = 'output/potential.txt'
data_form = pd.read_table(output_fname)
```

```python
title = data_form.columns.values
potential = data_form.values
[nx, ny, nz] = list(map(int, str.split(title[0])))
```

```python
layer = nz / 2
array = potential[int(layer * nx * ny) : int((layer + 1) * nx * ny)]
array = array.reshape(nx, ny)
plt.imshow(array)
plt.colorbar()
```

```python
ee = 1.6e-19
e0 = 8.8542e-12
layer = nz / 2

line_value = potential[int(layer * nx * ny + (ny / 2) * nx) : int(layer * nx * ny + (ny / 2 + 1) * nx)]
x = 1e-9 * np.array(range(-15, 15))

plt.figure(figsize=[8, 6])
x0 = 1e-10 * (np.array(range(-150, 150)) + 0.5)
plt.plot(x0, 10 * ee * np.exp(-np.abs(x0) / 2e-9) / np.abs(x0) / (4 * np.pi * e0), color=(1, 0.6, 0.6, 1))
plt.plot(x, line_value, marker='o', color=(1, 0, 0, 1), markersize=5, linewidth=0, mew=2)

plt.xlim([-1.5e-8, 1.5e-8])
plt.ylim([0, 40])
```
